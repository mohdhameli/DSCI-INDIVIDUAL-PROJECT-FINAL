### (1) Data description

I use two linked files: `players.csv` (one row per player) and `sessions.csv` (one row per play session), joined by an anonymized ID `hashedEmail`. `players.csv` contains age, gender, first name, self-reported experience level, total hours played, and newsletter subscription. `sessions.csv` records each sessionâ€™s start and end time as text and as server timestamps. From the timestamps I compute `session_minutes` for each session and, at the player level, `total_playtime_min`, `avg_session_min`, and `session_count`. Tables in the notebook report the number of observations and variables, numeric summaries (means to 2 d.p., plus range), and counts and percentages for all categorical variables.

Demographic and subscription data were entered on sign-up forms, while session times were logged automatically by the game server, so the data mix self-reported information with passively collected behaviour.

Main observable issues are missing or inconsistent timestamps (leading to missing or zero-length sessions), players with no sessions, and possible outliers in age or playtime. Categorical fields (gender, experience) may also contain rare or inconsistent labels. Less visible issues include selection bias (only logged players), measurement error in the server logs (e.g., crashes, AFK time), and unmeasured factors such as device, region, or motivation that may affect playtime but are not recorded.
